```python
# Que 4:

# Ans : 
#     DataFrame : It is a data structure that organizes data into a
#                 2-dimensional table of rows and columns.
#                 Whereas Series is a One-dimesional data structure and
#                 contians similar type of data.

# eg :
print("DataFrame : \n",df) # stores multiple data like name,age,gender
print("Series : \n",s) # stores single type of data -> like numbers
            
```

```python
# Que 7 :
s1 = pd.Series(["Pen","Pencil","Book","A4-Page"])
s2 = pd.Series([10,4,100,2])
s3 = pd.Series([100.150,200,400])

df = pd.DataFrame()
df["Product"] = s1
df["Price"] = s2
df["Count"] = s3
df
```
